## Feature Engineering

```python
# Creating additional features based on the above inference

# Transaction amount features
amount_threshold = np.log1p(1500)
fradulent_data['is_low_amount'] = np.where(fradulent_data['Transaction Amount'] < amount_threshold, 1, 0)

# Age features
fradulent_data['is_fraud_risk_age_group'] = np.where((fradulent_data['Customer Age'] >= 25) & (fradulent_data['Customer Age'] <= 40), 1, 0)

# Account registration features
fradulent_data['is_new_account'] = np.where(fradulent_data['Account Age Days'] < 40, 1, 0)

# Transaction time features
fradulent_data['is_early_morning'] = np.where((fradulent_data['Transaction Hour'] >= 0) & (fradulent_data['Transaction Hour'] <= 5), 1, 0)

# Interaction terms
fradulent_data['low_amount_and_new_account'] = fradulent_data['is_low_amount'] * fradulent_data['is_new_account']
fradulent_data['early_morning_and_high_risk_age'] = fradulent_data['is_early_morning'] * fradulent_data['is_fraud_risk_age_group']
```

```python
# List of high-risk locations
high_risk_locations = fradulent_data[fradulent_data['Is Fraudulent'] == 1]['Customer Location'].value_counts().head(20).index

# Binary flag for high-risk locations
fradulent_data['is_high_risk_location'] = fradulent_data['Customer Location'].apply(lambda x: 1 if x in high_risk_locations else 0)

# Fraud ratio by location
fraud_ratio = fradulent_data.groupby('Customer Location')['Is Fraudulent'].mean().to_dict()
fradulent_data['location_fraud_ratio'] = fradulent_data['Customer Location'].map(fraud_ratio)

# One-hot encoding for top 20 high-risk locations
fradulent_data['customer_location_grouped'] = fradulent_data['Customer Location'].apply(lambda x: x if x in high_risk_locations else 'Other')
fradulent_data = pd.get_dummies(fradulent_data, columns=['customer_location_grouped'], prefix='loc_grp', drop_first=True)

# Interaction terms
fradulent_data['high_risk_and_low_amount'] = fradulent_data['is_high_risk_location'] * fradulent_data['is_low_amount']
fradulent_data['high_risk_and_new_account'] = fradulent_data['is_high_risk_location'] * fradulent_data['is_new_account']
```

```python
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score
def displayMetrics(y_test, preds): 
    precision, recall, f_beta, support = precision_recall_fscore_support(y_test, preds, beta=5, pos_label=1, average='binary')
    auc = roc_auc_score(y_test, preds)
    accuracy = accuracy_score(y_test, preds)
    print(f"Accuracy is: {accuracy:.2f}")
    print(f"Precision is: {precision:.2f}")
    print(f"Recall is: {recall:.2f}")
    print(f"Fscore is: {f_beta:.2f}")
    print(f"AUC is: {auc:.2f}")
```
